```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
prob2_data =pd.read_csv("/Users/lisaricanek/BINF 6210/processed_cleveland_data.csv")
prob3_data = prob2_data.apply(pd.to_numeric, errors="coerce")
prob3_data.fillna(0, inplace=True)
X = prob3_data
```

```python
features = X.drop('diagnosis_of_heart_disease', axis=1)

# Standardize the features
scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)

# Perform k-means clustering (assuming k=3 for example)
kmeans = KMeans(n_clusters=3, random_state=42)
X['cluster'] = kmeans.fit_predict(features_standardized)

# Display the clustered data
print(X[['diagnosis_of_heart_disease', 'cluster']])

# Visualize the clusters (for 2D data, you can choose two features for plotting)
plt.scatter( X['resting_blood_pressure'],X['cholestoral'], c=X['cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Age')
plt.ylabel('Cholestoral')
plt.show()
```
